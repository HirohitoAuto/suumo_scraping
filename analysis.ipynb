```python
import pandas as pd

filename = "result/fukuoka_city_formatted.csv"
df = pd.read_csv(filename)
df.head()
```

### Filter

```python
# Calculate the threshold values
lower_threshold = df["price"].quantile(0.025)
upper_threshold = df["price"].quantile(0.975)

# Filter out the rows
df_filtered = df[(df["price"] > lower_threshold) & (df["price"] < upper_threshold)]
```

### 価格 vs. 築年数

```python
import plotly.express as px

fig = px.scatter(df_filtered, x="age", y="price", trendline="ols")
fig.update_layout(
    title="価格 vs. 築年数", xaxis_title="築年数[年]", yaxis_title="価格[万円]"
)
fig.show()
```

### 価格 vs. 駅までの時間

```python
fig = px.scatter(df_filtered, x="minutes", y="price", trendline="ols")
fig.update_layout(
    title="価格 vs. 駅までの時間", xaxis_title="駅までの時間[min]", yaxis_title="価格[万円]"
)
fig.show()
```

### 面積 vs. 築年数

```python
fig = px.scatter(df_filtered, x="age", y="area", trendline="ols")
fig.update_layout(
    title="価格 vs. 駅までの時間", xaxis_title="駅までの時間[min]", yaxis_title="価格[万円]"
)
fig.show()
```
